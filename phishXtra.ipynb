```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df1.corr()
plt.figure(figsize=(25, 13))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
```

Among some of the selected features high correlation is seen. Some tree-based models like Random Forest or XGBoost are less sensitive to
highly correlated features compared to linear models, as absence of multicollinearity is not assumed unlike linear models. In these models(ensemble methods) where feature importance is derived from feature interactions redundant or higly correlated features might still contribute valuable interactions.

```python
sns.barplot(data=df1, y="qty_slash_url", hue="phishing")
```

```python
# Contingency table
contingency_table = pd.crosstab(df1['qty_slash_url'], df1['phishing'])
print(contingency_table)
```

Here, qty_slash_url(number of'/' in a URL) and directory_length(no of directory characters) are the two most important features based on the chosen model.

On the above it can be seen that, with more no of slashes in a URL, the chances of a website being a phishing one increases. If it is >10 then definitely it is a phishing website.

```python
sns.barplot(data=df1, y="directory_length", hue="phishing")
```

```python
# Contingency table
contingency_table = pd.crosstab(df1['directory_length'], df1['phishing'])
print(contingency_table)
```

On the above it can be seen that, with more no of directory characters in a URL, the chances of a website being a phishing one increases.If >2, the probability of being a phishing website is higher.
